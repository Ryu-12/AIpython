```python
import matplotlib.pyplot as plt
Month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(12,6)) #설정을 먼저 해야한다
plt.ylim(0, max(Rainfall)*1.1)
for i, value in enumerate(Rainfall):
  plt.text(i, value+10, str(value), ha='center', va='bottom')
plt.tight_layout()
plt.plot(Month, Rainfall, 'b.-.', label='Monthly Rainfall')
plt.legend()
plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.show()
```

```python
print(np.zeros((2, 3)))      # 2x3 크기의 0 배열
print(np.ones(5))            # 길이 5의 1 배열
print(np.arange(0, 10, 2))   # 0부터 10까지 2씩 증가
print(np.linspace(0, 1, 5))  # 0~1 사이를 5개로 나눈 배열
print(np.random.rand(3))     # 0~1 사이의 난수 3개
```

```python
#문제1: 1차원 배열, 5개의 요소, 0으로 초기화된 배열

arr = np.zeros(5)
print(arr)
```

```python
#문제2: 1부터 10까지의 숫자 중 짝수만 포함하는 배열을 생성하세요.

arr = np.arange(2, 11, 2)
print(arr)
```

```python
#문제3: [2, 4, 6] 배열과 [10, 20, 30] 배열을 각각 생성하고, 곱셈 연산을 수행한 결과를 출력하세요.

arr1 = np.array([2, 4, 6])
arr2 = np.array([10, 20, 30])
result = arr1 * arr2
print(result)
```

```python
#문제4: 위 배열에서 20보다 큰 요소만 필터링해서 출력하세요.
print(result[result>20])
```
